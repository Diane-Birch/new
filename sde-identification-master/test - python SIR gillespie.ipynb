```python
import numpy as np
import scipy
import matplotlib.pyplot as plt

from sir import SIRG
```

```python
## test the new class
random_state = 1

sir = SIRG(random_state = random_state+1)

y0 = np.zeros(2,)
y0[0] = 0.02;   # initial condition for y1 (I, infected species)
y0[1] = 0.00;   # initial condition for y2 (R, recovered species)

n_initials = 100
rng = np.random.default_rng(random_state+0)
    
# create random initial conditions
y0 = np.ones((n_initials,1)) @ rng.uniform(low=0,high=0.5, size=(1, 2))

# note that even though we simulate until time_max=15, the results stop before that (when y1 = 0 internally)
time, y = sir.simulate(y0, time_max=15, time_step=1e-3)

fig,ax = plt.subplots(1,5, figsize=(15,3),sharey=True)
for k in range(5):
    ax[k].plot(time[k],y[k][:,0], color='black', lineWidth=2);
    ax[k].plot(time[k],y[k][:,1], color='red', lineWidth=2);
    ax[k].set_xlabel("time")
    if k == 0:
        ax[k].set_ylabel("species")
    ax[k].set_title(f"sample #{k+1}")

fig.tight_layout()
```

```python
## test the new class with the MATLAB initial conditions

sir = SIRG(random_state = 1)

y0 = np.zeros(2,)
y0[0] = 0.02;   # initial condition for y1 (I, infected species)
y0[1] = 0.00;   # initial condition for y2 (R, recovered species)

fig,ax = plt.subplots(1,1, figsize=(5,4))
ax=[ax]
k=0
time, y = sir.simulate(y0, time_max=4, time_step=1e-3)

ax[k].plot(time,y[:,0], color='black', lineWidth=2);
ax[k].plot(time,y[:,1], color='red', lineWidth=2);
ax[k].set_xlabel("time")
if k == 0:
    ax[k].set_ylabel("species")
ax[k].set_ylim([0,.9])
ax[k].set_xlim([0,np.max(time)+.5])

fig.tight_layout()
```
